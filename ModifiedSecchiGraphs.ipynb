```python
# import necessary modules to run code
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sp
import numpy as np
import math
```

```python
#importing dataframe SECCHI
file_path = '/home/jovyan/EmmaIacono/194903-202105_Cast_nccsv.csv'
secchi= pd.read_csv(file_path, skiprows = 218, low_memory= False)
#file contains metadata row 1-219
#so we are skipping indices 0-218
```

```python
# secchi_years_of_interest1, we want only to include the years 1969-2007
secchi_years_of_interest1= secchi[(secchi['Year']>1968) & (secchi['Year']<2008)]
```

```python
# secchi_years_of_interest2, we want only to include the years 2008-2021
secchi_years_of_interest2= secchi[(secchi['Year']>2007) & (secchi['Year']<2022)]
```

```python
# set the x and y variables and the colors of the plot and regression line ##
chart = sns.regplot(x=secchi_years_of_interest1['Year'], y=secchi_years_of_interest1['Secchi'], scatter_kws={"color": "darkturquoise"}, 
            line_kws={"color": "red"}) 

reg_line_xarray = chart.get_lines()[0].get_xdata()
reg_line_yarray= chart.get_lines()[0].get_ydata()

reg_slope = (reg_line_yarray[1]-reg_line_yarray[0])/(reg_line_xarray[1]-reg_line_xarray[0])
reg_intercept = reg_line_yarray[0]- reg_slope*reg_line_xarray[0]
print('Slope: ', reg_slope)
print('Intercept: ', reg_intercept)

#### Calculate standard deviation ####

secchi1_std = np.std(secchi['Secchi'])
print('Secchi1 standard deviation: ', secchi1_std)

sample_size = len(secchi_years_of_interest1['Secchi'])
print('Sample size: ', sample_size)

secchi1_std_error = secchi1_std/math.sqrt(sample_size)
print('Standard error: ', secchi1_std_error)

# chart_text = 'y = ' + str(round(reg_slope, 2)) + 'x + ' + str(round(secchi1_std_error,2))
chart_text = 'y = ' + str(round(reg_slope, 2)) + ' +/- ' + str(round(secchi1_std_error,2)) + ' m , ' + 'n = ' + (str(sample_size))
chart.text(1969,57, chart_text, style='italic', fontdict={'size': 10, 'family': 'Sans-serif'})

chart.set_title('Secchi Disk Depth: 1969-2007', fontdict={'size': 15, 'weight': 'bold'})

chart.set_xlabel('Year', fontdict={'size': 12, 'family': 'Sans-serif', 'weight': 'heavy'})
chart.set_ylabel('Secchi Depth (meters)', fontdict={'size': 12, 'family': 'Sans-serif', 'weight': 'heavy'})

plt.show()
fig = chart.get_figure()
fig.savefig('./figures/secchi_disk_depth.png')
```

```python
print(secchi['Distance'].dropna(how='all'))
```

```python
secchi['Distance_with_*'] = secchi['Distance'] * -1.852
```

```python
# Create a new dataframe that contains all the columns where distance_with_* is less than equal to 150
df_inshore = secchi[(secchi['Distance_with_*']<=150)]

# Create a new dataframe that contains all the columns where distance_with_* is greater than 450
df_offshore = secchi[(secchi['Distance_with_*']>450)]

# Create a new dataframe that contains all the columns where distance_with_* is greater than 150 and less than/ equal to 450
cond1 = secchi['Distance_with_*']<=450
cond2 = secchi['Distance_with_*']>150
df_transition = secchi[cond1 & cond2]
```

```python
### Inshore plot

# Filtered data from years 2007-2021

filtered_data = df_inshore[df_inshore['Year']>2006]

filtered_data_start_year = int(filtered_data['Year'].iloc[0])
filtered_data_end_year = int(filtered_data['Year'].iloc[-1])

plot_title = 'Inshore Secchi Depth: ' + str(filtered_data_start_year) + '-' + str(filtered_data_end_year)

# here, we are changing the x axis to year instead of distance from shore
plot_x = filtered_data['Year']
plot_y = filtered_data['Secchi']

#### Calculate standard deviation ####

inshore_std = np.std(filtered_data['Secchi'])
print('Inshore standard deviation: ', inshore_std)

sample_size = len(filtered_data)
print('Sample size: ', sample_size)

inshore_std_error = inshore_std/math.sqrt(sample_size)
print('Standard error: ', inshore_std_error)

chart = sns.regplot(x= plot_x, 
                    y= plot_y, 
                    scatter_kws={"color": "mediumaquamarine"}, 
                    line_kws={"color": "darkgreen"})

reg_line_xarray = chart.get_lines()[0].get_xdata()
reg_line_yarray= chart.get_lines()[0].get_ydata()

reg_slope = (reg_line_yarray[1]-reg_line_yarray[0])/(reg_line_xarray[1]-reg_line_xarray[0])
reg_intercept = reg_line_yarray[0]- reg_slope*reg_line_xarray[0]
print('Slope: ', reg_slope)
print('Intercept: ', reg_intercept)

# chart_text = 'y = ' + str(round(reg_slope, 4)) + 'x + ' + str(round(inshore_std_error,2))
chart_text = 'y = ' + str(round(reg_slope, 4)) + ' +/- ' + str(round(inshore_std_error,2)) + ' m , ' + 'n = ' + (str(sample_size))

chart.text(2007,37, chart_text, style='italic', fontdict={'size': 10, 'family': 'Sans-serif'})

chart.set_title(plot_title, fontdict={'size': 15, 'weight': 'bold'})
chart.set_xlabel('Year', fontdict={'size': 12, 'family': 'Sans-serif', 'weight': 'heavy'})
chart.set_ylabel('Secchi Depth (meters)', fontdict={'size': 12, 'family': 'Sans-serif', 'weight': 'heavy'})

plt.show()

fig = chart.get_figure()
fig.savefig('./figures/Year_inshore_secchi_depth.png')
```

```python
### Transition plot

# Filtered data from years 2007-2021

filtered_data = df_transition[df_transition['Year']>2006]

filtered_data_start_year = int(filtered_data['Year'].iloc[0])
filtered_data_end_year = int(filtered_data['Year'].iloc[-1])

plot_title = 'Transition Secchi Depth: ' + str(filtered_data_start_year) + '-' + str(filtered_data_end_year)

# here, we are changing the x axis to year instead of distance from shore
plot_x = filtered_data['Year']
plot_y = filtered_data['Secchi']

#### Calculate standard deviation ####

transition_std = np.std(filtered_data['Secchi'])
print('Transition standard deviation: ', transition_std)

sample_size = len(filtered_data)
print('Sample size: ', sample_size)

transition_std_error = transition_std/math.sqrt(sample_size)
print('Standard error: ', transition_std_error)



##### Plot #######

chart = sns.regplot(x= plot_x, 
                    y= plot_y, 
                    scatter_kws={"color": "darkslategrey"}, 
                    line_kws={"color": "turquoise"})

reg_line_xarray = chart.get_lines()[0].get_xdata()
reg_line_yarray= chart.get_lines()[0].get_ydata()

reg_slope = (reg_line_yarray[1]-reg_line_yarray[0])/(reg_line_xarray[1]-reg_line_xarray[0])
reg_intercept = reg_line_yarray[0]- reg_slope*reg_line_xarray[0]
print('Slope: ', reg_slope)
print('Intercept: ', reg_intercept)

chart_text = 'y = ' + str(round(reg_slope, 4)) + ' +/- ' + str(round(transition_std_error,2)) + ' m , ' + ' n = ' + (str(sample_size))

chart.text(2007,45, chart_text, style='italic', fontdict={'size': 10, 'family': 'Sans-serif'})

chart.set_title(plot_title, fontdict={'size': 15, 'weight': 'bold'})
chart.set_xlabel('Year', fontdict={'size': 12, 'family': 'Sans-serif', 'weight': 'heavy'})
chart.set_ylabel('Secchi Depth (meters)', fontdict={'size': 12, 'family': 'Sans-serif', 'weight': 'heavy'})

plt.show()
fig = chart.get_figure()
fig.savefig('./figures/Year_transition_secchi_depth.png')
```
